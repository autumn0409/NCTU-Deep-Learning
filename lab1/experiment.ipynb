```python
from generate_data import generate_linear, generate_XOR_easy
from show_result import show_result
from utils import to_label, cal_accuracy
from net import FC_Net
import matplotlib.pyplot as plt
from utils import to_label, cal_accuracy
```

```python
def plot_learning_curves(title, var, history, y_label):
    for i in range(len(var)):
        label=f'{var[i]}'
        plt.plot(history[i]['epochs'], history[i][y_label], label=label)
        
    plt.xlabel('Epoch')
    plt.ylabel(y_label.capitalize())
    plt.legend(title=title)
    plt.show()
```

```python
def plot_learning_curve(epochs, loss):
    plt.plot(epochs, loss, 'C1')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
```

```python
x_train_linear, y_train_linear = generate_linear()
x_train_xor, y_train_xor = generate_XOR_easy()
epochs=50000
```

# Different Learning Rate

```python
learning_rates=[1, 0.1, 0.01, 0.001]
```

```python
# train linear
linear_history = []
y_hat = None
for learning_rate in learning_rates:
    net_linear = FC_Net(4, 4)
    linear_history.append(net_linear.train(x_train_linear, y_train_linear, epochs=epochs, learning_rate=learning_rate))
    if learning_rate == 0.1:
        y_hat = net_linear.test(x_train_linear)
```

```python
print(y_hat)
y_hat_label = to_label(y_hat)
show_result(x_train_linear, y_train_linear, y_hat_label)
print(f'Accuracy = {cal_accuracy(y_train_linear, y_hat)}')
```

```python
plot_learning_curve(linear_history[1]['epochs'], linear_history[1]['loss'])
```

```python
plot_learning_curves('learning rate', learning_rates, linear_history, 'loss')
plot_learning_curves('learning rate', learning_rates, linear_history, 'accuracy')
```

```python
# train XOR
xor_history = []
y_hat = None
for learning_rate in learning_rates:
    net_XOR = FC_Net(10, 10)
    xor_history.append(net_XOR.train(x_train_xor, y_train_xor, epochs=epochs, learning_rate=learning_rate))
    if learning_rate == 0.1:
        y_hat = net_XOR.test(x_train_xor)
```

```python
print(y_hat)
y_hat_label = to_label(y_hat)
show_result(x_train_xor, y_train_xor, y_hat_label)
print(f'Accuracy = {cal_accuracy(y_train_xor, y_hat)}')
```

```python
plot_learning_curve(xor_history[1]['epochs'], xor_history[1]['loss'])
```

```python
plot_learning_curves('learning rate', learning_rates, xor_history, 'loss')
plot_learning_curves('learning rate', learning_rates, xor_history, 'accuracy')
```
